# Cleaning Data

```python
null_counts = df.isnull().sum()

print("Null counts in each column:")
print(null_counts)
```

```python
# Position column had an extra space at the end, stripped any extra from any other columns as well
df.columns = df.columns.str.strip()
```

```python
# Cleaning 'Yearly bonus + stocks in EUR', wasn't fully containing numbers so they're swapped to a number if the string contained the person's bonus
replace_func = lambda x: '0' if not str(x).replace('.', '').isdigit() else x

df['Yearly bonus + stocks in EUR'] = df['Yearly bonus + stocks in EUR'].apply(replace_func)

df['Yearly bonus + stocks in EUR'] = df['Yearly bonus + stocks in EUR'].astype(float)

empty_int_columns = ['Yearly bonus + stocks in EUR']

df[empty_int_columns] = df[empty_int_columns].fillna('0')

df[472:473]
```

```python
# Function to extract numbers from a string

def extract_number(text):
    if pd.isna(text):
        return None
    
    if isinstance(text, str):
        #Returns the first number found
        numbers = re.findall(r'\d+\.\d+|\d+', text) 
        return float(numbers[0]) if numbers else None  


# Apply function to 'Number of vacation days' to extract numbers from non-number answers
df['Number of vacation days'] = df['Number of vacation days'].apply(extract_number)

df['Number of vacation days'] = df['Number of vacation days'].astype(float)

mean_value_vacation_days = df['Number of vacation days'].mean()

# Fill null values 'Number of vacation days' with the calculated mean
df['Number of vacation days'] = df['Number of vacation days'].fillna(mean_value_vacation_days)

df
```

```python
# Fill null 'Age' values with the calculated mean
mean_value_age = df['Age'].mean()

df['Age'] = df['Age'].fillna(mean_value_age)
```

```python
# Fill null values in the 'Gender' column with the mode

#Decided to limit genders to 2 to improve speed and simplicity
mode_gender = df['Gender'].mode()[0]  
df['Gender'] = df['Gender'].fillna(mode_gender)

df.head(10)
```

```python
# Function to replace low frequencys, under 10, in columns with 'other'  

def replace_low_freq(d, col, threshold=10, replacement='other'):
    # Makes everything lower
    d[col] = d[col].str.lower()
    value_counts = d[col].value_counts()
    to_remove = value_counts[value_counts <= threshold].index
    tmp = d[col].replace(to_replace=to_remove, value=replacement)
    return tmp
```

```python
# Fill null values in the 'Position' column with the mode and replace low frequencys 

mode_position = df['Position'].mode()[0]  
df['Position'] = df['Position'].fillna(mode_position)
df['Position'] = replace_low_freq(df, 'Position', 5, "Other")
df['Position'].value_counts()
```

```python
# Change columns into floats
df['Total years of experience'] = df['Total years of experience'].apply(extract_number)
df['Total years of experience'] = df['Total years of experience'].astype(float)
```

```python
# Change columns into floats
df['Years of experience in Germany'] = df['Years of experience in Germany'].apply(extract_number)
df['Years of experience in Germany'] = df['Years of experience in Germany'].astype(float)
```

```python
# Fill null values in the 'Position' column with the mode and replace low frequencys 
mode_Seniority_level = df['Seniority level'].mode()[0]  
df['Seniority level'] = df['Seniority level'].fillna(mode_Seniority_level)

df['Seniority level'] = replace_low_freq(df, 'Seniority level', 5, "Other")
df['Seniority level'].value_counts()
```

For 'Total years of experience' and 'Years of experience in Germany' I thought it would be more accurate if we took the average amount of years for each 'Seniority level' instead of just taking the overall avearge years of experience for all workers.   

```python
# Finding average years for each seniority level for 'Total years of experience'
df['Seniority level'] = df['Seniority level'].str.lower()

years_experience_total_avg = df.groupby('Seniority level', as_index=False)['Total years of experience'].mean()
print(years_experience_total_avg)
```

```python

average_years_per_seniority = df.groupby('Seniority level')['Total years of experience'].transform('mean')

# Fill missing values in 'Total years of experience'
df['Total years of experience'] = df['Total years of experience'].fillna(average_years_per_seniority)
```

```python
# # Finding average years for each seniority level for 'Years of experience in Germany'
df['Seniority level'] = df['Seniority level'].str.lower()

years_experience_germany_avg = df.groupby('Seniority level', as_index=False)['Years of experience in Germany'].mean()
print(years_experience_germany_avg)
```

```python

average_years_per_seniority = df.groupby('Seniority level')['Years of experience in Germany'].transform('mean')

# Fill missing values in 'Years of experience in Germany'
df['Years of experience in Germany'] = df['Years of experience in Germany'].fillna(average_years_per_seniority)
```
